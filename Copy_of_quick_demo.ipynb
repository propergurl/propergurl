```python
# Upload Image

import os
from google.colab import files
import shutil

train_data_dir = "/content/SadTalker/examples/source_image"
os.makedirs(train_data_dir, exist_ok=True)
upload_folder = '/content/SadTalker/examples/source_image'

if os.path.isdir(upload_folder):
    shutil.rmtree(upload_folder)
os.mkdir(upload_folder)

############################################################

uploaded = files.upload()
for filename in uploaded.keys():
  dst_path = os.path.join(upload_folder, filename)
  print(f'move {filename} to {dst_path}')
  shutil.move(filename, dst_path)

############################################################

def main():

    upload_folder = "/content/SadTalker/examples/source_image"
    UBAH_NAMA_FOTO = "coba" #@param {type:'string'}

    for count, filename in enumerate(os.listdir(upload_folder)):
        dst =f"{UBAH_NAMA_FOTO}.png"
        src =f"{upload_folder}/{filename}"
        dst =f"{upload_folder}/{dst}"
        os.rename(src, dst)

if __name__ == '__main__':
    main()

############################################################

from IPython.display import clear_output
clear_output()
print("✅ Upload gambar sukses!")
############################################################
```

```python
# borrow from makeittalk
import ipywidgets as widgets
import glob
import os
import matplotlib.pyplot as plt

paths = 'examples/source_image'
entries = [os.path.splitext(filename)[0] for filename in os.listdir(paths)]
string_list = map(str, entries)
result = ''.join(string_list)

print("Choose the image name to animate: (saved in folder 'examples/')")
img_list = glob.glob1(paths, '*.png')
img_list.sort()
img_list = [item.split('.')[0] for item in img_list]
default_head_name = widgets.Dropdown(options=img_list, value= f'{result}')
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        plt.imshow(plt.imread(paths+'/{}.png'.format(default_head_name.value)))
        plt.axis('off')
        plt.show()
default_head_name.observe(on_change)
display(default_head_name)
plt.imshow(plt.imread(paths+'/{}.png'.format(default_head_name.value)))
plt.axis('off')
plt.show()
```
